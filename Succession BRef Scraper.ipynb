```python
#scrape to 2020
df = pd.DataFrame()
for i in range(2001,2020):    
    url = 'https://www.basketball-reference.com/leagues/NBA_'+str(i)+'_coaches.html'
    r = requests.get(url)
    soup = bs4.BeautifulSoup(r.text, 'lxml')

    table = soup.find(id="NBA_coaches")
    rows = table.find_all('tr')

    rows = iter(rows)
    header_1 = [td.text for td in next(rows).find_all('td') if td.text]
    header_2 = [td.text for td in next(rows).find_all('td') if td.text]
    header = header_1[:2] + header_2 + header_1[-2:]
#     print(header)
    for row in rows:
#         data = [td.text for td in row.find_all('td') if td.text]
        data = [d.text.strip() for d in row.find_all('td')]
        a = pd.Series(data)
        a = a.append(pd.Series(str(i)),ignore_index=True)
        df = df.append(a, ignore_index=True)
```

```python
# Get coaches
coaches = []
for i in range(2001,2020):    
    url = 'https://www.basketball-reference.com/leagues/NBA_'+str(i)+'_coaches.html'
    r = requests.get(url)
    soup = bs4.BeautifulSoup(r.text, 'lxml')

    table = soup.find(id="NBA_coaches")
    coachrows = table.find_all('th')
    rows = iter(coachrows)
    for row in rows:
        test = r'(?<=\.html">).+?(?=\</a)'
        fin = re.findall(test,str(row))
        #print(len(fin))
        if len(fin) >= 1:
            coaches.append(fin)
        else: 
            pass
        
```

```python
# coachrows
```

```python
coachesflat = [item for sublist in coaches for item in sublist]
coach_series = pd.Series(coachesflat)
coach_series
```

```python
df_cleanteam.insert(0,'coach',coach_series.to_numpy())
```

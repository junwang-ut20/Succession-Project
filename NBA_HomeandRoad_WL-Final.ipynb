```python
#start year (2019-20)
d = webdriver.Chrome(executable_path=path_to_chromedriver)
d.get('https://stats.nba.com/teams/traditional/?sort=TEAM_NAME&dir=-1&Season=2019-20&SeasonType=Regular%20Season&Location=Home')
d.find_element_by_xpath('/html/body/main/div[2]/div/div[2]/div/div/nba-stat-table/div[2]/div[1]').click()
table = d.find_element_by_class_name('nba-stat-table__overflow')
team_id = []
team_names = []
team_stats = []

for line_id, lines in enumerate(table.text.split('\n')):
    if line_id == 0:
        column_names = lines.split(' ')
    if line_id == 1:
        column_names2 = lines.split(' ')
    else:
        if line_id % 3 == 1:
            team_stats.append([float(i) for i in lines.split(' ')])
        if line_id % 3 == 2:
            team_id.append(lines)
        if line_id % 3 == 0:
            team_names.append(lines)
            
all_cols = column_names + column_names2
names_df = pd.DataFrame(team_names[1:],columns=['TEAM'])
names_df['YEAR'] = '2019-20'
stats_df = pd.DataFrame(team_stats,columns=all_cols[1:])
df_home = names_df.merge(stats_df,left_index=True,right_index=True)
```

```python
df_home
```

```python
#final looping HOME tables
for i in seasons:
    try:
        d = webdriver.Chrome(executable_path=path_to_chromedriver)
        d.get('https://stats.nba.com/teams/traditional/?sort=TEAM_NAME&dir=-1&Season='+str(i)+'&SeasonType=Regular%20Season&Location=Home')
        d.find_element_by_xpath('/html/body/main/div[2]/div/div[2]/div/div/nba-stat-table/div[2]/div[1]').click()
        table = d.find_element_by_class_name('nba-stat-table__overflow')

        team_id = []
        team_names = []
        team_stats = []

        for line_id, lines in enumerate(table.text.split('\n')):
            if line_id == 0:
                column_names = lines.split(' ')
            if line_id == 1:
                column_names2 = lines.split(' ')
            else:
                if line_id % 3 == 1:
                    team_stats.append([float(i) for i in lines.split(' ')])
                if line_id % 3 == 2:
                    team_id.append(lines)
                if line_id % 3 == 0:
                    team_names.append(lines)

        all_cols = column_names + column_names2
        names_df = pd.DataFrame(team_names[1:],columns=['TEAM'])
        names_df['YEAR'] = str(i)
        stats_df = pd.DataFrame(team_stats,columns=all_cols[1:])
        df_add = names_df.merge(stats_df,left_index=True,right_index=True)
        df_home = df_home.append(df_add)
    except:
        pass
```
